## Introduction and Motivation 


**[Q1]** The main motivation for the paper focuses on broadening our understanding of the labor supply relationship and how it impacts the distribution of work hours in the economy. This paper also analyzes the psychological benefits of having a job and of having low values of $z$. **[Q2]** Overall, the authors investigate the marginal value of non-work time and estimates the opportunity cost of unemployed applicants working an additional 5 hours. **[Q3]** Estimating the opportunity cost is difficult, however it allows the authors to estimate how much people value marginal leisure time.

The authors use a field experiment to estimate $Z$ (flow value of non-work time relative to worker's marginal product). **[Q7]** They posted a data entry position online and asked applicants to choose between two job bundles with different weekly hours, increasing 5 hours at a time, and randomly assigned wages. This allows researchers to estimate the Marginal Value of Time (MVT) over h. MVT increases in hours worked and creates the labor supply function. 

**[Q8]** $MVT_{i h}$ is the marginal value of non work time for individual "i" with number of hours worked "h". Here the h indicates the job with more hours. An alternative way to interpret $MVT_{i h}$ is the approximation of individual ***i's*** inverse labor supply function between long and short hours. 

The paper also focuses on estimating the flow value of non-work time relative to market wage $\tilde z$. This is calculated by combining MVT estimates with predicted market wages.**[Q4]** $Z$ and $\tilde Z$ are different. $z$ measures the flow value of non-work time relative to a workerâ€™s marginal product. A high $z$ follows the trend of the business cycle. $\tilde z$ estimates the flow value of non work time relative to market wage. This measurement takes MVT estimates and combines it with predicted wages. 

This paper relates to existing research on reservation wage such as Feldstein and Poterba (1984) and Roulet (2019), but here, the authors measure MVT at different hours of work in a natural environment. This paper also contributes to the existing literature of using field experiments to analyze labor supply relationships. These existing papers include Burtless and Hausman (1978), Angrist et al (2017).



## Experimental Design

**[Q5]** This fascinating 2017 experiment was conducted during the hiring process of telephone interviewer and data entry jobs and used cross-sectional survey data to complete the analysis. **[Q6], [Q7]** The applicants to these jobs are primarily women with varying levels of education from high school to college, and most are in the age range of 26-39. Advertisements for the positions were posted nationally in 80 major cities on a US job search site. The advertisements themselves only indicate the type of work being done and that compensation will be hourly: there is no indication on the advertisements at this time about the number of work hours nor the wage, and does not specify the job is remote.

The application was linked through the ad to an external website, where applicants first created an account and filled out a demographic questionnaire. Applicants were then shown two descriptions of job positions which differed only in the numbers of hourly work, and hourly wage. Job seekers were then asked to choose one of the two jobs to apply to. The framing of the two position descriptions were as follows:

*This is a one-month work-from-home [type] position.
The position is [X] hours per week.
This position pays [Y] dollars per hour*

*This is a one-month work-from-home [type] position.
The position is [X+5] hours per week.
This position pays [Z] dollars per hour*

Where [type] is the position type (telephone interviewer or data entry work), and [Y] and [Z] are dollar amounts. Each applicant was shown either both data entry positions or both phone survey positions. 

The additional earnings an applicant would receive from working the longer job instead of the shorter job were randomly varied. **[Q10]** The authors then take this opportunity to define the effective wage as $e_{ih} = \frac{h_i x w_{ih} - (h_i - 5)x w_{ih-5}}{5}$.

The authors randomly select $e_{ih} \in$ {0,4,8,12,16,18,20,24,28}. For the three lowest choices of $h$ (5 vs. 10, 10 vs. 15, and 15 vs. 20), the authors also let $e_{ih}$ take the values of {2,6,10}. This is because piloting showed the workers' values of time were lower at these hours levels and including them could lead to less extrapolation.

The authors forge on to mention that the randomly-chosen effective wage determined both jobs' hourly wage. **[Q10]** That is, the higher hourly job wage always paid 18 dollars/hr, and the other hourly wage was determined using the equation above. For example, if a 25-hour job paid $18/hr, then the 30-hour job would pay 15 dollars/hr such that wages were equivalent. 

**[Q11]** The authors then use these values of $e_{ih}$ to calculate $MVT_{ih}$ and therefore $\overline{MVT_h}$. **[Q9]** $\overline{MVT_{h}}$ is the average marginal hourly value of time in the sample. $\overline{MVT_{h}}$ is the expectation of $MVT_{ih}$. This allows us to estimate the value of one more hour of work on average to everyone included in the sample. The estimates of $\overline{MVT_{h}}$ are then combined for different hours levels up to 40 hours per week to form the estimate of $\tilde{z}$. 

$$\tilde{z} = \frac{1}{\hat{w}} [\frac{5(\overline{MVT_{5}} + \overline{MVT_{10}} + . . . + \overline{MVT_{40}}}{40}]$$

**[Q12]**  Where $\hat{w}$ is average predicted pretax hourly market wage, calculated using gender, race, education, and other data in the applicants' five-year age bin from the 2016 CPS MORGs survey. For unemployed applicants, the resulting wage is discounted by 6%, which is the estimated wage penalty employed hourly workers who separated from their job in the last 3 years.  The authors also use the CPS MORGs survey data to calculate $\hat{w}$,the average predicted pretax hourly market wage.

The authors also offered positions with overtime, which is greater than 40 hrs per week. The FLSA requires 1.5 times the regular wage in compensation per hour of overtime worked (the specifications of how this differs from hourly wage is discussed in "Estimation" section). Thus, for randomly selected participants, the following descriptions were shown:

*This is a one-month work-from-home [type] position.
The position is 40 hours per week.
This position pays $18 per hour.*

Then, for 45 and 50 hour positions:

*This is a one-month work-from-home [type] position.
The position is [45/50] hours per week.
This position pays $18 per hour for the first 40 hours and [EW] dollars per hour for the remaining [5/10] hours.*

Where EW is the effective wage. The authors randomly select EW $\in$ {27,29,31,33,35}. 

Due to concerns that it may be easier for applicants to make choices only when marginal wages differ than when average wages differ where they must determine effective wages, a validation study assessed whether applicants make similar decisions when choices are over marginal vs. average wages. Similar results were found between the two. 


## II Conceptual and Econometric framework 

### Conceptual framework 

Following the experimental design, applicants are shown a shorter and a longer job with hours and wages are given by $\left\{h_{i}-5, w_{i h-5}\right\}$ and $\left\{h_{i}, w_{i h}\right\}$, respectively.
The key concept is the $M V T_{i h}$ which is mean hourly value of time between $h-5$ and $h$ weekly hours of work for each individual $i$.
Authors present $M V T_{i h}$ as an approximation to $i$ 's inverse labor supply function between $h_{i}-5$ and $h_{i}$ hours relying on the theoretical concepts that work hours should be equal to the wage with the shadow value of time (Becker 1965; Gronau 1977; Heckman 1974). This means that $w_{i}\left(h_{i}-5\right)$ $<M V T_{i h}<w_{i}\left(h_{i}\right)$ if an individual's inverse labor supply function, $w_{i}(\cdot)$, is strictly increasing.
In other words one would choose to work strictly between $h_{i}-5$ and $h_{i}$ hours per week at a wage of $M V T_{i h}$.


The explicit assumption on the individual choices is provided: one would choose the longer job if the money she earns from working the additional five hours $\left(5 e_{i h}\right)$ exceeds the cost of working these additional hours $\left(5 M V T_{i h}\right)$. That is, if $M V T_{i h}<e_{i h}$.  **[Q9]** Hence, following these conceptual framework authors estimate the average marginal value of time ,  $\overline{M V T_{h}}=E\left[M V T_{i h}\right]$ by using applicants' choices over the $h$ and $h-5$ hour. After that they form $\tilde{z}$ estimate, the average flow value of non-work time, by combining estimates of $\overline{M V T_{h}}$ for different hours levels up to 40 hours per week to .


### Econometric framework and Estimation

**[Q11]** Assuming a logistic distribution for $M V T_{i h}$ authors estimate the probability of choosing the longer job $\operatorname{Pr}\left(\textbf{1}\left[h_{i}\right]=1 \mid e_{i h}\right)=\operatorname{Pr}\left(M V T_{i h}<e_{i h} \mid e_{i h}\right)$ by using a logit model for which the dependent variable is the indicator for choosing the longer position $\left(\mathbf{1}\left[h_{i}\right]\right)$ and the the explanatory variable is the effective wage $\left(e_{i h}\right)$. 

**[Q11]** Since according to FLSA high effective wages should be higher for overtime workers, 1.5 times the hourly wage, authors modified the estimation procedure of $\overline{M V T_{h}}$ for the overtime treatments. They use extrapolated $\overline{M V T_{h}}$ using the estimates of intercept in the non-overtime treatments. This procedure requires to do following two initial steps:

1. For each hour less than 40, run logit model regression as described above,
2. Calculate Intercept for each hour category and create a new variable called "intercept" with total 7 non-missing observations (corresponding to the 7 unique hour categories) 

```stata
use experiment.dta if emp_emp == 0 & chose_long < ., clear
qui{
    gen intercept = .
egen tag = tag(hours_high)
foreach hrs in 10 15 20 25 30 35 40 {
logit chose_long effectivewage if hours_high == `hrs'                         // 1
    
    
replace intercept = (1/(1+exp(-_b[_cons]))) if tag == 1 & hours_high == `hrs' // 2
}
}
```

From the second step above we can see that the constraint $\operatorname{Pr}\left(\mathbf{1}\left[h_{i}\right]=1 \mid e_{i h}=0\right)$ is used for estimated intercepts. The reason why the authors use this technique is because it increases the precision in the 40 vs. 45-hour choice dramatically. (It has little effect on the estimates for the 45 vs. 50-hour choice where there is less extrapolation.) 

Authors use estimated intercepts at lower hour treatments to predict the intercept as a function of $h$. Hence, the next part is to predict and define intercept constraint for overtime treatments, based on the estimation from previous part, by following the listed steps:

3. Create new variable called `hours` by taking averages of $h-5$ and $h$

```stata
gen hours=hours_high - 2.5                    // 3
```

4. Run linear probability model on the estimated `intercept` at lower hour treatments (h < 40) to predict the intercept as a function of $h$


```stata
qui reg intercept hours if hours <= 40 & tag == 1     // 4
predict p1
```

5. Predict intercept for 45 and 50 hours based on mean value of predicted intercepts and define constraints for those categories. 

```stata
qui {
sum p1 if hours == 42.5                           // 5
global int_p1con45 = -ln(1/`r(mean)' - 1)
sum p1 if hours == 47.5
global int_p1con50 = -ln(1/`r(mean)' - 1)
constraint define 45 [_cons] = $int_p1con45
constraint define 50 [_cons] = $int_p1con50
    
    preserve // Need this for appendix Figure 6 
sum p1 if hours == 37.5
global int_p1con40 = -ln(1/`r(mean)' - 1)
constraint define 40 [_cons] = $int_p1con40
    restore
    }
```

Looking at the different methods of picking the intercept constraints provided in the Online Appendix, one can see that they give similar estimates. At the same time, authors argue that these adjustments do not affect estimates of $\tilde{z}$.


Using the constraints from the previous part, the logit model can be run for all categories (step 6). Additionally, we may keep the report format of authors where the matrix is used to construct the final table. **[Q11]** The code-chunk below runs 9 total logit models (for each hours category) and simultaneously, inputs the coefficients and number of observations into the matrix named `mvt` (step 7).

```stata
qui{
mat mvt = J(30,6,.)
global allregs = ""
local row = 1
                                                // 6
foreach hrs in 10 15 20 25 30 35 40 45 50 {          
if `hrs' <= 40              logit chose_long effectivewage if hours_high == `hrs'
if inlist(`hrs',45,50)      logit chose_long effectivewage if hours_high == `hrs', const(`hrs') 
eststo lgt`hrs'
global allregs = "$allregs lgt`hrs'" //necessary for the next chunk 
    
//mat mvt[`row',1] = `hrs'                        // 7                                             
mat mvt[`row',3] = e(N)
local row = `row' + 3
}
}
```

**[Q17]** The final estimate of $\overline{M V T_{h}}$ is $-1$ times the estimated intercept divided by the slope coefficient. To illustrate this by an example we can assume $MTV_{ih}$ has mean $\mu$ and variance $s^{2}\pi/3$, where $s$ is the scale parameter. Recall the probability of choosing the longer job in logit model is defined by following formula :

$$Pr(\textbf{1}[h_i] = 1) = Pr(MTV_{ih} < e_{ih}) = Pr(\frac{MTV_{ih}-\mu}{s}) = \Lambda(\frac{1}{s} e_{ih} - \frac{\mu}{s}) = \Lambda(\beta_1 e + \beta_0) = \frac{1}{1+e^{(\beta_1 e + \beta_0)}}$$

where $\Lambda(\cdotp)$ is the standard logistic function. Thus, the slope coefficient of the logistic function, $\beta_1$ is the inverse of the scale parameter and the intercept is $\beta_0 = -\frac{\mu}{s}$. This is reassuring, because the logit cdf is 0.5 when $e=-\frac{\beta_{0} }{ \beta_{1}}$ which is when $e=\mu$, the mean of $MTV_{ih}$.

8. After running the regression for each hours choice (described in the previous chunk), **[Q13]** they combine the parameter estimates and covariance matrices from all hours choices and calculate robust standard errors on MVT using the delta method. The code-chunk below shows the described procedure where `nlcom` command is used after estimation by the `suest` command to calculate standard errors according to the delta method.

```stata
qui{
    suest $allregs , vce(robust)                                  // 8      
    local row = 1
    foreach hrs in 10 15 20 25 30 35 40 45 50 {
    nlcom  (- [lgt`hrs'_chose_long]_cons / [lgt`hrs'_chose_long]effectivewage)
        
mat v = r(V)
mat mvt[`row',2] = r(b)
mat mvt[`row'+1,2] = v[1,1]^.5
local row = `row' + 3
local w`hrs' = "(- [lgt`hrs'_chose_long]_cons / [lgt`hrs'_chose_long]effectivewage)"

preserve // we need the following part for Fig 1
        collapse chose_long, by(hours_high effectivewage)
tempfile points
save `points'
restore
}
}
```

## Main Treatments 
The authors first visually summarize the data parametrically and non-parametrically. **[Q15]** **Non-parametric methods** do not require any assumptions regarding the pdf or distribution of the data. This methods is usually a weaker statistical procedure and require a larger sample size to offset this effect. On the other hand, **parametric methods** always hinge on pdf assumptions of the population and sometimes rely on a normality assumption (or, in this paper, logistic distribution). That is, they present actual values with the fitted line of logit model.

In ***Figure 1***, the authors create a scatterplot of the share of applicants who chose longer hours against effective wages for each hour choice. A logit MLE is also utilized to constrain the probabilities between zero and one and creates a cdf of the MVT distribution. **[Q16]** The figures illustrate that individuals only choose to work more hours if the wage is higher. There is a positive relationship between hours worked and effective wages. That is, taking into account their opportunity cost, people rationally require higher effective wage for more time spent. 

The code below generates variables for high hours, effective wage. Logit MLE is applied and we generate the `logitp` for each high hour. We then plot the share of applicants choosing longer hours across effective wage. Our figures exactly match that of the authors'. The interpretation remains the same. 

```stata
qui {
clear
set obs 9
gen hours_high = _n*5 + 5
expand 100
gen effectivewage = runiform(-.2,35)
gen logitp = .
foreach hh in 10 15 20 25 30 35 40 45 50 {
estimates restore lgt`hh'
predict a
replace logitp = a if hours == `hh'
drop a
}

append using `points', gen(points)

label define hrslab 10 "5 vs. 10 Hours" 15 "10 vs. 15 Hours" 20 "15 vs. 20 Hours" ///
25 "20 vs. 25 Hours" 30 "25 vs. 30 Hours" 35 "30 vs. 35 Hours" 40 "35 vs. 40 Hours" ///
45 "40 vs. 45 Hours" 50 "45 vs. 50 Hours"
label values hours_high hrslab

twoway (scatter chose_long effectivewage if points == 1, symbol(o) color(navy)) ///
(line logitp effectivewage if points == 0, sort color(navy) ///
ylabel(0(.5)1, labsize(small)) ymtick(0(.25)1, grid gmin gmax) xlabel(0(10)35, labsize(small)) xmtick(5(10)35) ///
subtitle(,fcolor(none) nobox) ///
ytitle(Share Choosing Longer Job) xtitle("Effective Wage")), ///
by(hours_high, graphregion(color(white)) ///
title(Figure 1. Marginal Value of Time) subtitle(Unemployed Applicants) legend(off) ///
yrescale xrescale note(`" "'))
}
```

# Discussion

There is no doubt leisure time and time spent at home is highly valued. The estimates found in this paper support that notion, despite the popular contrary belief that Americans live to work. However, to properly compare the estimated found in this paper to those found in prior literature, the authors must account for possible gaps between wages and marginal products, taxes, fixed costs of employment, and child-care costs.

_Wages vs. Marginal Products:_
 **[Q18]** workers' marginal products likely weakly exceed their wages, so estimates of $z$ based on marginal product are weakly smaller than those based on wages, $\tilde{z}$. In the monopsony framework, the percent difference between marginal product and wage is the inverse of the elasticity of labor supply to the firm. Assuming a high elasticity of around 4, which is in line with Manning's estimates from 2011, MP will be 25% higher than wage, which indicates $z$ will be 20% *lower* than $\tilde{z}$. 

_Taxes:_ 
$z$ is estimated using after-tax marginal product. Applying an effective tax rate of 0.2 increases the estimate of $z$ by 25%.

_Fixed Costs of Employment (commuting and forgone benefits):_
Assuming the average worker spends approximately one hour a day commuting, this would add 0.06 to the estimated value of $z$.
As far as forgone benefits, there was no significant change in $z$ to be found because unemployed applicants often are not receiving Unemployment Insurance benefits, despite being eligible.

# Conclusion and Assessment

This paper successfully maps out the relationship of effective wage and labor supply to better understand how people value their non-work time. It was found that labor supply is elastic at low hours of work, and relatively inelastic for full-time+ positions (high hours of work). Because of this, there are few part time workers in the United States. A point we wish the authors would have expanded on is how this varying elasticity of labor supply affects the productivity and economic growth of a country. Another potential weakness of this study was the overwhelming proportion of women to men. Women often have very different career goals than men, and having a more diversified sample could potentially make for more interesting results. A strength of this paper, however, is the inclusion of the fixed costs of employment in the discussion section. These costs, specifically child-care and commuting, certainly affects how people allocate and prioritize their time. However, the section regarding childcare costs seems haste and does not mention where the estimates of childcare costs come from, which left us feeling as though this section, too, could be expanded on.
